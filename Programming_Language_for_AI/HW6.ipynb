```python
class Bear:
    def eats(self):
        return 'berries'
class Rabbit:
    def eats(self):
        return 'clover'
class Octothorpe:
    def eats(self):
        return 'campers'
bear = Bear()
rabbit = Rabbit()
octothorpe = Octothorpe()

print("Bear eats:", bear.eats())
print("Rabbit eats:", rabbit.eats())
print("Octothorpe eats:", octothorpe.eats())
```

```python
class Laser:
    def does(self):
        return 'disintegrate'
class Claw:
    def does(self):
        return 'Crush'
class SmartPhone:
    def does(self):
        return 'ring'
class Robot:
    def __init__(self):
        self.laser = Laser()
        self.claw = Claw()
        self.smartphone = SmartPhone()
    def does(self):
        print("Robot can:")
        print("- " + self.laser.does())
        print("- " + self.claw.does())
        print("- " + self.smartphone.does())
robot = Robot()
robot.does()
```

```python
class Element:
    def __init__(self, name, symbol, number):
        self.__name = name
        self.__symbol = symbol
        self.__number = number

    @property
    def name(self):
        return self.__name
    @property
    def symbol(self):
        return self.__symbol
    @property
    def number(self):
        return self.__number

hydrogen = Element("Hydrogen", "H", 1)
print("Element name:", hydrogen.name)
print("Element symbol:", hydrogen.symbol)
print("Element number:", hydrogen.number)
```

```python
import math
class QuantumParticle:
    def __init__(self, a,b,c):
        self.__a = a
        self.__b = b
        self.__c = c
    
    def getA(self):
        return self.__a
    def getB(self):
        return self.__b
    def getC(self):
        return self.__c
    
    def getDiscriminant(self):
        return self.__b**2 - 4*self.__a*self.__c
    def getRoots1(self):
        discriminant = self.getDiscriminant()
        if discriminant < 0:
            return 0
        else:
            return (-self.__b + math.sqrt(discriminant)) / (2 * self.__a)
        
    def getRoots2(self):
        discriminant = self.getDiscriminant()
        if discriminant < 0:
            return 0
        else:
            return (-self.__b - math.sqrt(discriminant)) / (2 * self.__a)
        
a=float(input("Enter coefficient a: "))
b=float(input("Enter coefficient b: "))
c=float(input("Enter coefficient c: "))

equation = QuantumParticle(a, b, c)
discriminant = equation.getDiscriminant()
if discriminant > 0:
    print("The equation has two distinct real roots")
    print("Root 1:", equation.getRoots1())
    print("Root 2:", equation.getRoots2())
elif discriminant == 0:
    print("The equation has one real root")
    print("Root:", equation.getRoots1())
else:
    print("The equation has no roots")
```

```python
import time
class StopWatch:
    def __init__(self):
        self.__starttime = time.time()
        self.__endtime = self.__starttime
    
    def getStartTime(self):
        return self.__starttime
    
    def getEndTime(self):
        return self.__endtime 

    def start(self):
        self.__starttime = time.time()
    def stop(self):
        self.__endtime = time.time()
    
    def get_elapsed_time(self):
        return int((self.getEndTime() - self.getStartTime()) * 1000)
    
stop__watch = StopWatch()
stop__watch.start()
total_sum = 0
for i in range(1, 1000001):
    total_sum += i
stop__watch.stop()
print("Elapsed time in milliseconds:", stop__watch.get_elapsed_time())
```
