```python
import numpy as np
A = np.random.randint(1, 31, size=(3, 3))
print("a:")
print(A)

B = np.full((3, 3), 2)
print("b:")
print(B)

C = np.arange(1, 13).reshape(2, 6)
print("c:")
print(C)

print("d:")
print(A * B)

print("e:")
try:
    e = np.dot(A, C)
    print(e)
except ValueError as e:
    print("Error in matrix multiplication:", e)

mean = np.mean(A)
print("f:")
print("mean A:", mean)

max_A = np.max(A)
print("g:")
print("max A:", max_A)

std_A = np.std(A)
print("h:")
print("std A:", std_A)

flatten_A = A.flatten(order='C')
print("i:")
print("flattened A:", flatten_A)

max_A_axis0 = np.max(A, axis=0)
print("j:")
print("max A along axis 0:", max_A_axis0)
```

```python
import numpy as np

A = np.arange(3,13)
print("a:")
print(A)

B = np.arange(10, 0, -1)
print("b:")
print(B)

A = A.reshape(2, 5)
print("c:")
print(A)

B = B.reshape(5, 2)
print("d:")
print(B)

e = np.concatenate((A, B.reshape(2, 5)), axis=1)
print("e:")
print(e)

f = np.concatenate((A.reshape(5, 2), B), axis=0)
print("f:")
print(f)

g = np.vsplit(e, 2)
print("g:")
print("ndarray1:", g[0])
print("ndarray2:", g[1])

h = np.hsplit(f, 2)
print("h:")
print("ndarray1:", h[0])
print("ndarray2:", h[1])

i = A.flatten(order='F')
print("i:")
print(i)

j = B.flatten(order='C')
print("j:")
print(j)
```

```python
import numpy as np
arr = np.random.randint(1, 101, size=100)

median = np.median(arr)
print("a:")
print("Median:", median)

variance = np.var(arr)
print("b:")
print("Variance:", variance)

max_idx = np.argmax(arr)
print("c:")
print("Index of maximum value:", max_idx)

min_idx = np.argmin(arr)
print("d:")
print("Index of minimum value:", min_idx)

condition_index = np.where((arr > 35) & (arr <= 90))[0]
print("e:")
print("Indices of values between 35 and 90:", condition_index)
```
