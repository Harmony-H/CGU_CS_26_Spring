```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, (ax1, ax2) = plt.subplots(2, 1)

scatter_artist = ax1.scatter(x, y1, label='Sin', color='tab:blue')
legend_artist1 = ax1.legend()

ax1.set_title('Sin and Cos Curves')
ax1.set_ylabel('Y-axis')

line_artist = ax2.plot(x, y2, label='Cos', color='tab:green')
legend_artist2 = ax2.legend()

ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(1, 501, 500)
y = np.random.randint(1, 501, 500)

fig, ax = plt.subplots()

ax.scatter(x, y, alpha=0.7, color='tab:blue')

ax.set_title('Scatter Plot of 500 Random Points (1 to 500)')
ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data1 = np.random.randint(1, 3001, 500)
data2 = np.random.randint(1, 3001, 1000)

fig, ax = plt.subplots()
ax.boxplot([data1, data2], tick_labels=['Group 1 (N=500)', 'Group 2 (N=1000)'])
ax.set_title('Box Plot of Two Random Datasets (1-3000)')
ax.set_ylabel('Values')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

N = 10000
x = np.random.randint(1, 20001, N)
y = np.random.randint(1, 20001, N)
z = np.random.randint(1, 20001, N)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z, s=2, alpha=0.5, color='tab:blue')
ax.set_title('3D Scatter Plot of 10000 Random Points (1-20000)')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randint(1, 101, 10000)
bins = range(0, 101, 10)

fig, ax = plt.subplots(figsize=(10, 6))
n, bins_edges, patches = ax.hist(data, bins=bins, edgecolor='black')
colors = plt.cm.tab10.colors
for i, patch in enumerate(patches):
    patch.set_facecolor(colors[i])
ax.set_title('Histogram of 10000 Random Integers (1-100)')
ax.set_xlabel('Value Range (Bins)')
ax.set_ylabel('Frequency')
ax.set_xticks(bins)
plt.show()
```
