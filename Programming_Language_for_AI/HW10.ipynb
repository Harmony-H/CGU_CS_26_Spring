```python
import pandas as pd

df = pd.read_excel('dataset.xlsx')

condition = (df['AGE'] >= 50) & (df['AGE'] <= 60)
filtered_df = df[condition]

result = filtered_df[['HEIGHT', 'WEIGHT']].agg(['mean', 'std'])
print(result)
```

```python
import pandas as pd

df = pd.read_excel('dataset.xlsx')

condition = (df['AGE'] >= 30) & (df['AGE'] <= 50) & (df['SEX'] == 0)
filtered_df = df[condition]

result = filtered_df[['HEIGHT', 'WEIGHT']].agg(['mean', 'std'])
print(result)
```

```python
import pandas as pd

df2 = df.copy()
condition = (df2['AGE'] >= 40) & (df2['AGE'] <= 50) & (df2['SEX'] == 0)
df2.loc[condition, 'WEIGHT'] = None
none_count = df2['WEIGHT'].isna().sum()
print(f"體重為 None 的總人數為：{none_count} 人")
```

```python
import pandas as pd

df_clean = df.dropna()

condition = (df_clean['AGE'] >= 30) & (df_clean['AGE'] <= 70) & (df_clean['SEX'] == 0)
filtered_df = df_clean[condition]

pearson_corr = filtered_df['WEIGHT'].corr(filtered_df['AGE'], method='pearson')
spearman_corr = filtered_df['WEIGHT'].corr(filtered_df['AGE'], method='spearman')

print(f"Pearson correlation:{pearson_corr:.4f}")
print(f"Spearman correlation:{spearman_corr:.4f}")
```
