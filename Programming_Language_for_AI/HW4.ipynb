```python
num1 = int(input("Enter first integer: "))
num2 = int(input("Enter second integer: "))
gcd = 1
for i in range(1, min(num1, num2) + 1):
    if num1 % i == 0 and num2 % i == 0:
        gcd = i
print(f"The greatest common divisor for {num1} and {num2} is {gcd}.")
```
