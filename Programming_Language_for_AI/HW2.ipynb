```python
import math
x1,y1,x2,y2,x3,y3 = eval(input("Enter three points:"))

a = math.sqrt((x3-x2)**2+(y3-y2)**2)
b = math.sqrt((x3-x1)**2+(y3-y1)**2)
c = math.sqrt((x1-x2)**2+(y1-y2)**2)

A = round(math.degrees(math.acos((a*a-b*b-c*c)/(-2*b*c))),2)
B = round(math.degrees(math.acos((b*b-a*a-c*c)/(-2*a*c))),2)
C = round(math.degrees(math.acos((c*c-a*a-b*b)/(-2*a*b))),2)
print("The three angles are %f %f %f" %(A,B,C))
```

```python
import math
x1,y1,x2,y2=eval(input("Enter the latitude and longitude of two points in degrees:"))
x1 = math.radians (x1)
y1 = math.radians (y1)
x2 = math.radians (x2)
y2 = math.radians (y2)
d = 6371.018*math.acos(math.sin(x1)*math.sin(x2)+math.cos(x1)*math.cos(x2)*math.cos(y1-y2))
print(f"The great circle distance of two point is {d:.2f}km.")
```

```python
letter = """Dear {salutation} {name},

Thank you for your letter. We are sorry that our {product}
{verbed} in your {room}. Please note that it should never
be used in a {room}, especially near any {animals}.

Send us your receipt and {amount} for shipping and handling.
We will send you another {product} that, in our tests,
is {percent}% less likely to have {verbed}.

Thank you for your support.

Sincerely,

{spokesman}
{job_title}"""

salutation = "Mr."
name = "Chen"
product = "toaster"
verbed = "exploded"  
room = "kitchen"
animals = "cats"
amount = "$50"
percent = 20
spokesman = "John Doe"
job_title = "Customer Service Manager"

result = letter.format(
    salutation=salutation, 
    name=name, 
    product=product, 
    verbed=verbed, 
    room=room, 
    animals=animals, 
    amount=amount, 
    percent=percent, 
    spokesman=spokesman, 
    job_title=job_title
)

print(result)
```
