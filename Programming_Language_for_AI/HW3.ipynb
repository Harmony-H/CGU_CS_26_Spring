```python
list1 = input("Enter numbers separated by spaces: ")
list1 = list1.split()
list2 = [] 
for item in list1:
    num = int(item)
    if num not in list2:
        list2.append(num)
print(list2)
```

```python
list = input("Enter integers between 1 and 100: ")
list = list.split()
list = [int(x) for x in list]
list.sort()
for i in range(len(list)):
    if i > 0 and list[i-1] == list[i]:
        continue
    occurrences = list.count(list[i])
    if(occurrences > 1):
        print(f"{list[i]} occurs {occurrences} times")
    else:
        print(f"{list[i]} occurs {occurrences} time")
```

```python
list = input("Enter scores: ")
list = list.split()
list = [int(x) for x in list]
max_score = max(list)
for i in range(len(list)):
    if list[i] >= max_score - 10:
        print(f"Student {i} score is {list[i]} and grade is A")
    elif list[i] >= max_score - 20:
        print(f"Student {i} score is {list[i]} and grade is B")
    elif list[i] >= max_score - 30:
        print(f"Student {i} score is {list[i]} and grade is C")
    elif list[i] >= max_score - 40:
        print(f"Student {i} score is {list[i]} and grade is D")
    else:
        print(f"Student {i} score is {list[i]} and grade is F")
```

```python
fruits_prices = {"apple": 0.80, "grape": 0.60, "orange": 0.40}
veggies_prices = {"tomato": 1.80, "pepper": 1.40, "onion": 1.23}
order = {"grape": 4, "tomato": 10, "orange": 4, "pepper": 1}
all_prices = {**fruits_prices, **veggies_prices}
total_bill = 0.0
print("--- Order Breakdown ---")
for item, quantity in order.items():
    if item in all_prices:
        item_total = quantity * all_prices[item]
        total_bill += item_total
        print(f"{item.capitalize():<6}: ${all_prices[item]:.2f} x {quantity:2} = ${item_total:.2f}")
print(f"Total: ${total_bill:.2f}")
```
