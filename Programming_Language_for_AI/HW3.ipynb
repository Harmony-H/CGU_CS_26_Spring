```python
numbers = input("Enter integers between 1 and 100: ")
numbers = numbers.split()
numbers = [int(x) for x in numbers]
counts = {}

for num in numbers:
    if num in counts:
        counts[num] += 1
    else:
        counts[num] = 1

for num in sorted(counts):
    if counts[num] > 1:
        print(f"{num} occurs {counts[num]} times")
    else:
        print(f"{num} occurs {counts[num]} time")
```

```python
list = input("Enter scores: ")
list = list.split()
list = [int(x) for x in list]
max_score = max(list)
for i in range(len(list)):
    if list[i] >= max_score - 10:
        print(f"Student {i} score is {list[i]} and grade is A")
    elif list[i] >= max_score - 20:
        print(f"Student {i} score is {list[i]} and grade is B")
    elif list[i] >= max_score - 30:
        print(f"Student {i} score is {list[i]} and grade is C")
    elif list[i] >= max_score - 40:
        print(f"Student {i} score is {list[i]} and grade is D")
    else:
        print(f"Student {i} score is {list[i]} and grade is F")
```

```python
import re
from collections import Counter

def analyze_text_file(filename):
    try:
        with open("3_4_input.txt", 'r', encoding='utf-8') as file:
            text = file.read().lower()
    except FileNotFoundError:
        print(f"Error: Could not find the file '{filename}'.")
        return

    words = re.findall(r"[a-z0-9]+", text)
    word_counts = Counter(words)
    total_words = len(words)
    print(f"(1) Total # of words: {total_words}")

    total_types = len(word_counts)
    print(f"(2) Total # of word's type: {total_types}")

    top_three_words = word_counts.most_common(3)
    print(f"(3) The first three words that most frequently appear and their #:")
    for word, count in top_three_words:
        print(f"    {word}: {count}")

    words_appearing_once = [word for word, count in word_counts.items() if count == 1]
    words_appearing_once.sort() 
    first_three_unique = words_appearing_once[:3]
    print(f"(4) The first three words that only appear once (in lexical order):")
    print(f"    {', '.join(first_three_unique)}")

    print("(5) Character counts:")
    valid_chars = [char for char in text if char.isalnum() and char.isascii()]
    char_counts = Counter(valid_chars)

    for char in sorted(char_counts.keys()):
        print(f"{char}->{char_counts[char]}")

analyze_text_file('3_4_input.txt')
```

```python
fruits_prices = {"apple": 0.80, "grape": 0.60, "orange": 0.40}
veggies_prices = {"tomato": 1.80, "pepper": 1.40, "onion": 1.23}
order = {"grape": 4, "tomato": 10, "orange": 4, "pepper": 1}
all_prices = {**fruits_prices, **veggies_prices}
total_bill = 0.0
print("--- Order Breakdown ---")
for item, quantity in order.items():
    if item in all_prices:
        item_total = quantity * all_prices[item]
        total_bill += item_total
        print(f"{item.capitalize():<6}: ${all_prices[item]:.2f} x {quantity:2} = ${item_total:.2f}")
print(f"Total: ${total_bill:.2f}")
```
