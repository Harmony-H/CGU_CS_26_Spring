```python
import os
print("a. 目前目錄的檔案:")
current_dir_files = os.listdir('.')
for item in current_dir_files:
    print(item)

print("b. 上一層目錄的檔案:")
parent_dir_files = os.listdir("..")
for item in parent_dir_files:
    print(item)

print("c. 建立並寫入 test.txt:")
test1 = "This is a test of the emergency text system"
with open("test.txt", "w", encoding="utf-8") as f:
    f.write(test1)

print("d. 讀取 test.txt 並比對內容:")
with open("test.txt", "r", encoding="utf-8") as f:
    test2 = f.read()

print(f"test1: {test1}")
print(f"test2: {test2}")

if test1 == test2:
    print("結果：test1 與 test2 完全一樣！")
else:
    print("結果：test1 與 test2 不一樣。")
```

```python
import re

with open("mammoth.txt", "r", encoding="utf-8") as f:
    text = f.read()

words_start_c = re.findall(r"\bc\w*\b", text, re.IGNORECASE)
words_start_c_unique = list(dict.fromkeys(words_start_c))
print("a. c 開頭的所有單字 (去除重複):")
print(words_start_c_unique)

words_start_c_4 = re.findall(r"\bc\w{3}\b", text, re.IGNORECASE)
words_start_c_4_unique = list(dict.fromkeys(words_start_c_4))
print("b. c 開頭的四字母單字 (去除重複):")
print(words_start_c_4_unique)

words_end_r = re.findall(r"\b\w*r\b", text, re.IGNORECASE)
words_end_r_unique = list(dict.fromkeys(words_end_r))
print("c. 以 r 結束的所有單字 (去除重複):")
print(words_end_r_unique)
print("-" * 50)

words_3_vowels = re.findall(r"\b\w*[aeiou]{3}\w*\b", text, re.IGNORECASE)
words_3_vowels_unique = list(dict.fromkeys(words_3_vowels))
print("d. 包含連續三個母音的單字 (去除重複):")
print(words_3_vowels_unique)
```

```python
def analyze_file(filename):
    try:
        with open(filename, "r", encoding="utf-8") as f:
            text = f.read()
        char_count = len(text)
        words = text.split()
        word_count = len(words)
        lines = text.splitlines()
        line_count = len(lines)

        print(f"總字元數: {char_count}")
        print(f"總單字數: {word_count}")
        print(f"總行數　: {line_count}")

    except FileNotFoundError:
        print(f"錯誤：找不到名為 '{filename}' 的檔案。")

analyze_file("mammoth.txt")
```
