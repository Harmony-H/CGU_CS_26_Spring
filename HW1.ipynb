```python
population = 3120333422
newborn = 365*24*60*60/7
death = 365*24*60*60/13
immigrant = 365*24*60*60/45
for i in range(5):
    population = population + newborn - death + immigrant
    print("The population in", i+1, "years is", int(population))
```

```python
import time
total_seconds = int(time.time())
current_second = total_seconds % 60
total_minutes = total_seconds // 60
current_minute = total_minutes % 60
total_hours = total_minutes // 60
current_hour = total_hours % 24
taipei_hour = (current_hour + 8) % 24

print(f"{taipei_hour:02}:{current_minute:02}:{current_second:02}")
```
